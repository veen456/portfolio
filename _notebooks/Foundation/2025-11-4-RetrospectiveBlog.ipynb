**Class Lessons**

In our class lessons, we learned about many coding concepts in javascript including variables, functions,  iterations, classes and methods.


Variable: A variable is a storage container that holds information like a number or a string.
You can define a variable through diffeent keywords:

**Var:** Outdated and not used anymore
 
**Let:** The most common way to define a variable

**Const:** Define a variable where it stays the same the whole time

```python
// x, y and z are all variables
var x = 1
let y = 2
const z = 3
```

Functions: Functions are an efficient way to define a sequence of events without repeatedly writing it when you need it

Functions are defined by the keyword `function`

```python
function double(a){
    let new_a = a * 2
    console.log("You number doubles is " + new_a);
}

double(5);
```

 Iteration: You can use loops to go through a series of data quickly, there are multiple kinds of loops (for loops and while loops)


```python
let i = 10
for (i>=0){
    i--;
    console.log(i)
}

let n = 10
while (n >= 0){
    n--;
    console.log(n)
}
```
